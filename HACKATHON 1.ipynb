```python
# ==========================================
# MOVIE RECOMMENDATION SYSTEM
# Team G
# ==========================================

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import NearestNeighbors

# ==========================================
# SAMPLE MOVIE DATASET
# ==========================================

movies = pd.DataFrame({

    "Movie":[
        "Avengers",
        "Batman",
        "Titanic",
        "Frozen",
        "Joker",
        "Interstellar",
        "Avatar",
        "Inception",
        "Toy Story",
        "Spider-Man",
        "Iron Man",
        "Doctor Strange"
    ],

    "Genre":[
        "Action",
        "Action",
        "Romance",
        "Animation",
        "Drama",
        "Sci-Fi",
        "Sci-Fi",
        "Sci-Fi",
        "Animation",
        "Action",
        "Action",
        "Sci-Fi"
    ],

    "Rating":[
        9.2,
        8.7,
        8.5,
        8.3,
        9.0,
        9.1,
        8.8,
        9.3,
        8.1,
        8.9,
        8.8,
        8.7
    ],

    "Popularity":[
        98,
        94,
        91,
        88,
        95,
        96,
        93,
        97,
        86,
        94,
        92,
        90
    ]

})

print("Movie Dataset\n")
print(movies)

# ==========================================
# ENCODE GENRE
# ==========================================

encoder = LabelEncoder()

movies["Genre_Code"] = encoder.fit_transform(movies["Genre"])

# ==========================================
# FEATURES
# ==========================================

X = movies[["Genre_Code","Rating","Popularity"]]

# ==========================================
# TRAIN MODEL
# ==========================================

model = NearestNeighbors(n_neighbors=3)

model.fit(X)

print("\nModel Trained Successfully!")

# ==========================================
# USER INPUT
# ==========================================

print("\nAvailable Genres")

for g in encoder.classes_:
    print("-",g)

genre = input("\nEnter Genre : ")

rating = float(input("Minimum Rating (1-10): "))

if genre not in encoder.classes_:
    print("Genre Not Found")
else:

    genre_code = encoder.transform([genre])[0]

    user = [[genre_code,rating,95]]

    distance,index = model.kneighbors(user)

    print("\nRecommended Movies\n")

    for i in index[0]:
        print("---------------------------")
        print("Movie :",movies.iloc[i]["Movie"])
        print("Genre :",movies.iloc[i]["Genre"])
        print("Rating :",movies.iloc[i]["Rating"])
        print("Popularity :",movies.iloc[i]["Popularity"])
```
